#fonksiyonun içine argümanları bir liste içerisinde vermek gerekiyor ve genişlik ve uzunluk belirmek gerekiyor.[ sol başlangıç,alt başlangıç,genişlik,uzunluk]

```python
benimFigure = plt.figure() #boş çerçeve oluşturur
figureAxes = benimFigure.add_axes([0.8,0.8,0.3,0.3])
figureAxes.plot(numpyDizisi1,numpyDizisi2,"g")
figureAxes.set_xlabel("X ekseni veri ismi")
figureAxes.set_ylabel("Y ekseni veri ismi")
figureAxes.set_title("Grafik Başlığı")
```

```python
figure2 = plt.figure()
eksen1 = figure2.add_axes([0.1,0.1,0.9,0.9])
eksen2 = figure2.add_axes([0.3,0.5,0.3,0.3])

eksen1.plot(numpyDizisi1,numpyDizisi2,"g")
eksen1.set_xlabel("X Ekseni")
eksen1.set_xlabel("X Ekseni")
eksen1.set_ylabel("Y Ekseni")
eksen1.set_title("Ana Grafik Başlık")


eksen2.plot(numpyDizisi2,numpyDizisi1,"b")
eksen2.set_xlabel("X Ekseni")
eksen2.set_xlabel("X Ekseni")
eksen2.set_ylabel("Y Ekseni")
eksen2.set_title("Küçük Grafik Başlık")
```

```python
(benimFigure,benimEksenler) = plt.subplots(nrows =1,ncols=2) 
#figure ve eksen dönderir : aslında tuple var
type(benimEksenler) #bu bir dizi old. için for loop içinde yapmak gerek
for eksen in benimEksenler:
    eksen.plot(numpyDizisi1,numpyDizisi2,"g")
    eksen.set_xlabel("X ekseni")
    
plt.tight_layout() #iki grafik arasını ayarlar
#benimFigure
```

```python
yeniFigur = plt.figure(dpi =100)
#figsize=(3,3):figure boyutunu doğrudan fonk.içinde verdik eğer vermezsen de eksen büyüklüğüne göre boyut ayarlanıyor.
# dpi =100: dpi verirsen de pikselini ayarlar, dpi daha sonra 

yeniEksen = yeniFigur.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyDizisi1,numpyDizisi1**2,label = "numpyDizisi ** 2")
yeniEksen.plot(numpyDizisi1,numpyDizisi1**3,label = "numpyDizisi **3")
yeniEksen.legend(loc =2)  
#etiket koymaya yarar yer, loc ile yer değiştiriyor
```

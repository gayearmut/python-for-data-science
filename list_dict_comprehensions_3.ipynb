Amaç key'i string, valuesu aşağıdaki gibi sözcük oluşturmak.

sadece sayısal değişkenler için yapmak istiyoruz.

```python
import seaborn as sns
df = sns.load_dataset("car_crashes")
df.columns
```

```python
num_col = [col for col in df.columns if df[col].dtype != "O"]

dic = {}
agg_list = ["mean","min","max","sum"]
```

```python
for col in num_col:
  dic[col] = agg_list
  #sol taraf değişken isimleri sağ taraf ise liste
```

```python
#kısayol

new_dictionary = {col: agg_list for col in num_col}
```

```python
print(new_dictionary)
```

```python
df[num_col].head()
```

```python
df[num_col].agg(new_dictionary)
```

```python
y = dataFrame["Type"].values
X= dataFrame.drop("Type",axis=1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler =MinMaxScaler( copy=True,feature_range=(0, 1))
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test =scaler.transform(X_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Activation,Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
X_train.shape
```

```python
model = Sequential()
model.add(Dense(units = 30,activation="relu"))
#30 tane nöron koy, neden: kaç tane kolon varsa giriş layerı o kadar olması önerilir
model.add(Dense(units = 15,activation="relu"))
model.add(Dense(units = 15,activation="relu"))
model.add(Dense(units = 1,activation="sigmoid"))
#aynı kural 30-1 arasında layer vermeyi öneriyor
#regresyondason layera(output) aktivasyon kodu koymamıştık ancak bu sefer 
#sigmoid aktivasyon kodunu koyacağız. s,gmoid 0-1 arası değer verir ve sınıflandırma problemlerinde
#output layerına bunu koyarak bunu elde edebiliyoruz

model.compile(loss ="binary_crossentropy",optimizer ="adam")
```

```python
model.fit(x = X_train,y=y_train,epochs =700,validation_data=(X_test,y_test),verbose=1)
```

```python
dataLoss= pd.DataFrame(model.history.history)
```

```python
dataLoss.plot()
```

overfitting olmuş, loss aşıı derece düşmüş ve val_loss da artmış. Bunları izleyip yorum yapabiliyoruz. Ör: val_loss saçma dercede artmaya başlarsa o zaman eposch u durdur ve train etme.

```python
model = Sequential()
model.add(Dense(units = 30,activation="relu"))

model.add(Dense(units = 15,activation="relu"))
model.add(Dense(units = 15,activation="relu"))
model.add(Dense(units = 1,activation="sigmoid"))

model.compile(loss ="binary_crossentropy",optimizer ="adam")
```

```python
earlyStopping =EarlyStopping(monitor="val_loss",mode="min", verbose=1,patience=25)
#val_loss: monitor:gözlemlenecek olan
#patience ,epochs lara bakıyor ve modeli gerekli durumda durdurur,yüksek verirsen geç durur
#verbose: ne kadar yazdıracak
#mode:min;val loss u düşük tutmaya çalışıyoruz bu yüzden min, ör; accurracy gözlemlesek max alırız
```

```python
model.fit(x=X_train, y=y_train, epochs=700, validation_data=(X_test, y_test), verbose=1, callbacks=[earlyStopping])
```

```python
dataLoss = pd.DataFrame(model.history.history)
```

```python
dataLoss.plot()
```

```python
model = Sequential()

model.add(Dense(units=30,activation = "relu"))
model.add(Dropout(0.6))

model.add(Dense(units=15,activation = "relu"))
model.add(Dropout(0.6))

model.add(Dense(units=15,activation = "relu"))
model.add(Dropout(0.6))

model.add(Dense(units=1,activation = "sigmoid"))

model.compile(loss="binary_crossentropy",optimizer = "adam")
```

```python
model.fit(x=X_train, y=y_train, epochs=700, validation_data=(X_test, y_test), verbose=1, callbacks=[earlyStopping])
```

```python
kayipDF = pd.DataFrame(model.history.history)
```
